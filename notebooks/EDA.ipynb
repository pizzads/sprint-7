```python
# Importar las librer√≠as para explorar con plotly
import pandas as pd
import plotly.express as px
```

```python
# Cargar el dataser
df = pd.read_csv('../vehicles_us.csv')
```

```python
# Vista general de los datos
df.head(10)
```

```python
# Informaci√≥n del dataset
df.info()
```

```python
# Estad√≠sticas generales
df.describe()
```

```python
# Ver cu√°ntos valores nulos hay por columna
df.isna().sum()
```

```python
# LIMPIEZA DE VALORES NULOS

# 'is_4wd': muchos valores faltantes probablemente indican que el veh√≠culo NO tiene tracci√≥n 4x4
df['is_4wd'].fillna(0, inplace=True)

# 'cylinders': el n√∫mero de cilindros es importante, y 0 no es v√°lido, as√≠ que usamos la moda (valor m√°s frecuente)
df['cylinders'].fillna(df['cylinders'].mode()[0], inplace=True)

# 'model_year': 0 no es un a√±o real, as√≠ que mejor usar la mediana para evitar distorsiones
df['model_year'].fillna(df['model_year'].median(), inplace=True)

# 'odometer': 0 podr√≠a malinterpretarse como veh√≠culo nuevo, as√≠ que usamos la mediana como valor representativo
df['odometer'].fillna(df['odometer'].median(), inplace=True)

# 'paint_color': el color no es esencial para el an√°lisis, pero rellenamos con 'unknown' para mantener consistencia
df['paint_color'].fillna('unknown', inplace=True)

# Confirmamos que no haya valores nulos
df.isna().sum()
```

```python
# Histograma de precios para mostrar c√≥mo se distribuyen los precios
px.histogram(df, x = 'price', nbins = 50, title = 'Distribuci√≥n de Precios de Veh√≠culos')
```

```python
# Grafica Scatter Plot: Precio vs A√±o del Modelo
px.scatter(df, x = 'model_year', y = 'price', title = 'Precio vs A√±o del Modelo')
```

```python
# Graficar Boxplot de precio por tipo de veh√≠culo para ver c√≥mo var√≠a el precio dependiendo del tipo de veh√≠culo
px.box(df, x = 'type', y = 'price', title = 'Precio por Tipo de Veh√≠culo')
```
