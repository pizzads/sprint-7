```python
# Importar las librerías para explorar con plotly
import pandas as pd
import plotly.express as px
```

```python
# Cargar el dataser
df = pd.read_csv('../vehicles_us.csv')
```

```python
# Vista general de los datos
df.head(10)
```

```python
# Información del dataset
df.info()
```

```python
# Estadísticas generales
df.describe()
```

```python
# Ver cuántos valores nulos hay por columna
df.isna().sum()
```

```python
# LIMPIEZA DE VALORES NULOS

# 'is_4wd': muchos valores faltantes probablemente indican que el vehículo NO tiene tracción 4x4
df['is_4wd'].fillna(0, inplace=True)

# 'cylinders': el número de cilindros es importante, y 0 no es válido, así que usamos la moda (valor más frecuente)
df['cylinders'].fillna(df['cylinders'].mode()[0], inplace=True)

# 'model_year': 0 no es un año real, así que mejor usar la mediana para evitar distorsiones
df['model_year'].fillna(df['model_year'].median(), inplace=True)

# 'odometer': 0 podría malinterpretarse como vehículo nuevo, así que usamos la mediana como valor representativo
df['odometer'].fillna(df['odometer'].median(), inplace=True)

# 'paint_color': el color no es esencial para el análisis, pero rellenamos con 'unknown' para mantener consistencia
df['paint_color'].fillna('unknown', inplace=True)

# Confirmamos que no haya valores nulos
df.isna().sum()
```

```python
# Histograma de precios para mostrar cómo se distribuyen los precios
px.histogram(df, x = 'price', nbins = 50, title = 'Distribución de Precios de Vehículos')
```

```python
# Grafica Scatter Plot: Precio vs Año del Modelo
px.scatter(df, x = 'model_year', y = 'price', title = 'Precio vs Año del Modelo')
```

```python
# Graficar Boxplot de precio por tipo de vehículo para ver cómo varía el precio dependiendo del tipo de vehículo
px.box(df, x = 'type', y = 'price', title = 'Precio por Tipo de Vehículo')
```
